**Kotimaan matkojen lukumäärä esitettynä vuosittain 2012-2021**

**mallina käytetään  K means -klusterointia**


```python

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
```

```python

df = pd.read_excel("YO_2_koko maa.xlsx")
df
```

```python

df.dtypes
```

```python

df.describe()
```

```python
#valitaan vain kotimaan matkailu df2 dataframeksi

df2 = df.iloc[:30,:]
df2.head()
```

```python
#poistetaan ikä-muuttuja df2:sta ja luodaan df3

df3 = df2.drop(columns = 'ikä')
df3.head()
```

```python
# df4 jossa vain vuosi ja matkojen lkm tuhansina - muuttujat

df4 = df3.iloc[:,2:]
df4.head()
```

**Normalisoidaan muuttujat vuosi ja matkojen lkm tuhansina, koska käytetään klusterointimallia.**

```python

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df4 = scaler.fit_transform(df4)

```

```python
# muodostetaan uudestaan df4 dataframe

df4 = pd.DataFrame(df4, columns =['vuosi', 'matkojen lkm tuhansina'])
df4.head()
```

```python

df4.describe()
```

```python

X = df4
X.head()
```

**Lasketaan Silhouette score, jolla saadaan klustereiden määrä.**

```python

#Silhouette score

from sklearn.metrics import silhouette_score

K = range(2, 10)

scores = [] #scoret sijoitetaan listaan

for k in K:

  kmeanModel = KMeans(n_clusters=k).fit(X)

  score = silhouette_score(X, kmeanModel.labels_, metric='euclidean')

  scores.append(score)

  print('k=',k,' silhouette ',score)
```

```python

scores
```

```python

import numpy as np

k_optimal = K[np.array(scores).argmax()]
k_optimal
```

**Silhouette score perusteella valitaan 2 klusteria, ja käytetään klusterointiin K-means mallia.**

```python

malli = KMeans(n_clusters=2)

malli.fit(X)

klusterit = malli.predict(X)
klusterit # eroteltu 2 klusteria toisistaan
```

```python

# lisätään df4:n uusi muuttuja, klusteri

df4['klusteri'] = klusterit
df4.head()
```

**Piirretään kuvaaja, jossa matkojen lukumäärä on klusteroitu vuosi-muuttujan perusteella 2 klusteriin.**

```python
#VUOSI X-AKSELILLA ja MATKOJEN LKM TUHANSINA Y-AKSELILLA

fig = plt.figure()
ax = fig.add_subplot()

#scatter plot
ax.scatter(X.iloc[:,0], X.iloc[:,1],c = klusterit)

ax.set_xlabel(X.columns[0])
ax.set_ylabel(X.columns[1])

plt.show()
```

***Kaksi klusteria muodostui sen perusteella, että eniten kotimaan matkoja tekivät ikäryhmä 25-64-vuotiaat ja vähemmän nuoret sekä seniorit.***

> ikäryhmässä 15-25 tehtiin kotimaan matkoja ikäryhmää 65-85 enemmän

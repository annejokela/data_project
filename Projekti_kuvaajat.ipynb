```python
#Kotimaan ja ulkomaan matkat vuosina 2012-2021

df2 = pd.read_excel("YO_0_linregr_04.xlsx")
df2.head()
```

```python

df2.describe()
```

```python

# MATKOJEN LUKUMÄÄRÄ KOTIMAASSA JA ULKOMAILLA VUOSINA 2012-2021
fig = px.line(df2, x='vuosi', y='matkat yön yli lkm tuhansia', color='matkakohde', markers=True)
fig.show()
```

```python

# YÖPYMISTEN LUKUMÄÄRÄ KOTIMAASSA JA ULKOMAILLA VUOSINA 2012-2021
fig = px.line(df2, x='vuosi', y='yöpymiset lkm tuhansia', color='matkakohde', markers=True)
fig.show()
```

```python
# MATKOJEN LUKUMÄÄRÄ ERI MATKAKOHTEISSA VUOSINA 2012-2022
fig = px.line(df, x='vuosi', y='Matkojen lkm tuhansina', color='Matkakohde', markers=True)
fig.show()
```

```python
# MATKOJEN LUKUMÄÄRÄ (TUHANSINA) VUOSINA 2012-2022, symbolin koko kertoo yöpymisten lukumäärän
fig = px.scatter(df,  x="vuosi", y="Matkojen lkm tuhansina", color="Matkakohde", size='Yöpymisten lkm tuhansina')
fig.show()
```

```python
# MATKOJEN LUKUMÄÄRÄ JA YÖPYMISTEN LUKUMÄÄRÄ ERI MATKAKOHTEISSA (vuosi näkyy hover-palkissa)
fig = px.scatter(df,  x="Matkojen lkm tuhansina", y="Yöpymisten lkm tuhansina", color="Matkakohde", size='vuosi')
fig.show()
```
